```
# 白話文：「我把東西安裝下來，程式裡有一些參數，我要先更改他才可以使用」
# 修改makefile檔案
# 將檔案中的OpenCV值改成1
# 且將GPU的值也改為1
# 且將CUDNN的值也改為1
# 便可以使用

# OpenCV 要為新版，新增了Mosaic這個 YOLOv4 所提出的Augmentation方法

%cd /drive/My\ Drive/app/darknet
!sed -i 's/OPENCV=0/OPENCV=1/' Makefile
!sed -i 's/GPU=0/GPU=1/' Makefile
!sed -i 's/CUDNN=0/CUDNN=1/' Makefile

!make
```
